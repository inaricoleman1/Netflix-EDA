# Data visualization

```python
plt.plot(netflix['release_year'])
plt.plot(netflix['type'])
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('type')
plt.ylabel('release_year')
 
plt.show()
```

```python
plt.plot(netflix['release_year'], netflix['type'], color='red', marker='o')
plt.title('type Vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Unemployment Rate', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
labels=['TV Show', 'Movie']
plt.pie(netflix['type'].value_counts().sort_values(),labels=labels,explode=[0.1,0.1],
        autopct='%1.2f%%',colors=['lightblue','royalblue'], startangle=90)
plt.title('Type of Netflix Content')
plt.axis('equal')
plt.show()
```

```python
plt.bar(netflix['release_year'], netflix['type'])
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('type')
plt.ylabel('release_year')
 
# Adding the legends
plt.show()
```

If we want to rotate the pie chart to a different angle, use this attribute with its appropriate value.

Syntax:
